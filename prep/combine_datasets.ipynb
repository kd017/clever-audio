```python
import os
import pandas as pd
```

```python
GRAMMY_WINNERS_ENRICHED = os.path.join('..', 'data', 'grammy-winners-enriched.csv')
SONGS_RELEASED_ENRICHED = os.path.join('..', 'data', 'songs-released-enriched.csv')
TOP_100_SONGS_ENRICHED = os.path.join('..', 'data', 'top-100-songs-enriched.csv')
COMBINED_DATA = os.path.join('..', 'data', 'combined-data.csv')
```

```python
grammy_winners_df = pd.read_csv(GRAMMY_WINNERS_ENRICHED)
grammy_winners_df.head()
```

```python
top_100_df = pd.read_csv(TOP_100_SONGS_ENRICHED)
top_100_df.head()
```

```python
songs_released_df = pd.read_csv(SONGS_RELEASED_ENRICHED)
songs_released_df.head()
```

```python
grammy_winners_df.loc[:, 'Winner'] = 1
grammy_winners_df.head()
```

```python
grammy_winners_df.shape
```

```python
grammy_winners_df.drop_duplicates(inplace=True)
grammy_winners_df.dropna(inplace=True)
```

```python
grammy_winners_df.shape
```

```python
winning_tracks = grammy_winners_df.TrackId.tolist()
winning_tracks
```

```python
len(winning_tracks)
```

```python
songs_released_df.shape
```

```python
songs_released_df.drop_duplicates(inplace=True)
songs_released_df.dropna(inplace=True)
```

```python
songs_released_df.shape
```

```python
songs_released_df[songs_released_df.Acousticness.isnull()]
```

```python
songs_released_non_winners_df = songs_released_df[~songs_released_df.TrackId.isin(winning_tracks)]
songs_released_non_winners_df.head()
```

```python
songs_released_non_winners_df.shape
```

```python
songs_released_non_winners_df.loc[:, 'Winner'] = 0
```

```python
songs_released_non_winners_df.columns
```

```python
songs_released_non_winners_df.Winner.value_counts()
```

```python
top_100_df.shape
```

```python
top_100_df.drop_duplicates(inplace=True)
top_100_df.dropna(inplace=True)
```

```python
top_100_df.shape
```

```python
top_100_non_winners_df = top_100_df[~top_100_df.TrackId.isin(winning_tracks)]
top_100_non_winners_df.head()
```

```python
top_100_non_winners_df.shape
```

```python
top_100_non_winners_df.loc[:, 'Winner'] = 0
```

```python
top_100_non_winners_df.columns
```

```python
top_100_non_winners_df.Winner.value_counts()
```

```python
non_winners_df = pd.concat([songs_released_non_winners_df, top_100_non_winners_df])
non_winners_df.head()
```

```python
non_winners_df.shape
```

```python
non_winners_df.drop_duplicates(inplace=True)
```

```python
non_winners_df.shape
```

```python
final_df = pd.concat([grammy_winners_df, non_winners_df])
final_df.head()
```

```python
final_df.shape
```

```python
final_df.to_csv(COMBINED_DATA, index=False)
```
