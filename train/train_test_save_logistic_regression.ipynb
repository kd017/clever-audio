```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
COMBINED_DATA = os.path.join('..', 'data', 'combined-data.csv')
```

```python
songs_df = pd.read_csv(COMBINED_DATA)
songs_df.head()
```

```python
songs_df.columns
```

```python
X = songs_df.drop(labels=['Artist', 'Title', 'URL', 'TrackId', 'Album', 'Image', 'Winner'], axis=1)
X.head()
```

```python
X.dtypes
```

```python
X.Explicit = X.Explicit.astype(int)
```

```python
X.Explicit.value_counts()
```

```python
#excluded_features = ['Year',  'Duration (ms)', 'Key',  'Mode', 'Speechiness', 'Tempo', 'Valence', 'Explicit']
excluded_features = ['Key']
X.drop(labels=excluded_features, axis=1, inplace=True)
```

```python
y = songs_df['Winner']
y.head()
```

```python
y.value_counts()
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
## Now Create the Logistic Regression Model
```

```python
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
import pickle

regression_model = LogisticRegression()
regression_model
```

```python
## Train model data
```

```python
regression_model.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {regression_model.score(X_train, y_train)}")
print(f"Testing Data Score: {regression_model.score(X_test, y_test)}")
```

```python
## Make Prediction
```

```python
predictions = regression_model.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
filename = os.path.join('..', 'model', 'grammy_prediction_logistic_model.h5')
pickle.dump(regression_model, open(filename, 'wb'))
```
